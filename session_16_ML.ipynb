```python
def print_confusion_matrix(tp, tn, fp, fn):
    print("Confusion Matrix")
    print("-" * 35)
    print("\t\tPredicted")
    print("\t\tPositive\tNegative")
    print(f"Actual Positive\t{tp}\t\t{fn}")
    print(f"Actual Negative\t{fp}\t\t{tn}")

# Example
print_confusion_matrix(50, 120, 10, 20)
```

```python
tp = 60
fp = 10
fn = 30

precision = tp / (tp + fp)
recall = tp / (tp + fn)

print("Precision =", round(precision, 2))
print("Recall =", round(recall, 2))
```

```python
tp = 40
tn = 150
fp = 20
fn = 10

precision = tp / (tp + fp)
recall = tp / (tp + fn)

f1_score = 2 * (precision * recall) / (precision + recall)

print("Precision =", round(precision, 2))
print("Recall =", round(recall, 2))
print("F1 Score =", round(f1_score, 2))
```
