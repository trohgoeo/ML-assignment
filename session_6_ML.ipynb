```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("/content/sample_data/mobile_phone_prices.csv")

# Features
X = df[["RAM", "Storage", "Camera_MP"]]

# Target
y = df["Price"]

# Train model
model = LinearRegression()
model.fit(X, y)

print("Regression Model Trained Successfully")
```

```python
print("Intercept:", model.intercept_)

coefficients = pd.DataFrame({
    "Feature": X.columns,
    "Coefficient": model.coef_
})

print(coefficients)
```

```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor

X_vif = df[["RAM", "Storage", "Camera_MP", "Battery_mAh"]]

vif = pd.DataFrame()

vif["Feature"] = X_vif.columns

vif["VIF"] = [
    variance_inflation_factor(X_vif.values, i)
    for i in range(X_vif.shape[1])
]

print(vif)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Remove Battery_mAh
X_new = df[["RAM", "Storage", "Camera_MP"]]

model2 = LinearRegression()

model2.fit(X_new, y)

prediction = model2.predict(X_new)

print("New Coefficients")

for feature, coef in zip(X_new.columns, model2.coef_):
    print(feature, ":", coef)

print("\nIntercept:", model2.intercept_)

print("\nR² Score:", r2_score(y, prediction))
```
