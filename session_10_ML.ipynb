```python
import numpy as np

# Sigmoid Function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Test Values
print("Sigmoid(-2) =", sigmoid(-2))
print("Sigmoid(0) =", sigmoid(0))
print("Sigmoid(3) =", sigmoid(3))
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression

# Dataset
data = {
    "likes": [120,250,80,500,300,60,150,700,90,450],
    "has_caption": [1,1,0,1,1,0,1,1,0,1],
    "viral": [0,1,0,1,1,0,0,1,0,1]
}

df = pd.DataFrame(data)

# Features
X = df[["likes", "has_caption"]]

# Target
y = df["viral"]

# Logistic Regression Model
model = LogisticRegression()

model.fit(X, y)

print("Model Coefficients:")
print(model.coef_)

print("Intercept:")
print(model.intercept_)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid Function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# X values
x = np.linspace(-10, 10, 200)

# Y values
y = sigmoid(x)

# Plot
plt.figure(figsize=(8,5))

plt.plot(x, y, color="blue", label="Sigmoid Function")

# Threshold Line
plt.axhline(y=0.5, color="red", linestyle="--", label="Threshold = 0.5")

plt.title("Sigmoid Function")

plt.xlabel("x")

plt.ylabel("Sigmoid(x)")

plt.legend()

plt.grid(True)

plt.show()
```

```python
# Classification Function
def classify(probability, threshold=0.5):

    if probability >= threshold:
        return "Viral"
    else:
        return "Not Viral"

# Test Probabilities
probabilities = [0.2, 0.6, 0.8]

print("Threshold = 0.5")

for p in probabilities:
    print(p, "->", classify(p, 0.5))

print()

print("Threshold = 0.7")

for p in probabilities:
    print(p, "->", classify(p, 0.7))
```
