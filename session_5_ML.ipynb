```python
import numpy as np
import matplotlib.pyplot as plt

# Instagram usage (hours per day for one week)
instagram_hours = np.array([1, 2, 3, 4, 5, 6, 7])

# Battery percentage remaining at end of day
battery_percentage = np.array([90, 85, 80, 72, 65, 58, 50])

# Scatter Plot
plt.figure(figsize=(7,5))
plt.scatter(instagram_hours, battery_percentage, color="blue", s=80)

plt.title("Instagram Usage vs Battery Percentage")
plt.xlabel("Instagram Usage (Hours)")
plt.ylabel("Battery Percentage (%)")

plt.grid(True)
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Feature (X)
X = instagram_hours.reshape(-1,1)

# Target (y)
y = battery_percentage

# Create Model
model = LinearRegression()

# Train Model
model.fit(X, y)

# Print Slope and Intercept
print("Slope (Coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)
```

```python
import matplotlib.pyplot as plt

# Predicted Values
y_pred = model.predict(X)

plt.figure(figsize=(7,5))

# Scatter Plot
plt.scatter(X, y, color="blue", s=80, label="Actual Data")

# Regression Line
plt.plot(X, y_pred, color="red", linewidth=2, label="Regression Line")

plt.title("Instagram Usage vs Battery Percentage")
plt.xlabel("Instagram Usage (Hours)")
plt.ylabel("Battery Percentage (%)")

plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

# Calculate MSE
mse = mean_squared_error(y, y_pred)

print("Mean Squared Error (MSE):", mse)
```
