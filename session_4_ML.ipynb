```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Load dataset
df = pd.read_csv("/content/sample_data/ipl_player_stats(2).csv")

# Select numeric columns
numeric_cols = ["runs", "wickets", "matches", "strike_rate"]

# Standard Scaling
scaler = StandardScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

# Save scaled dataset
df.to_csv("/content/sample_data/ipl_scaled(2).csv", index=False)

print(df)
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv("/content/sample_data/flipkart_products(2).csv")

print("Before Scaling")
print(df.min())
print(df.max())

scaler = MinMaxScaler()

scaled = scaler.fit_transform(df)

scaled_df = pd.DataFrame(scaled, columns=df.columns)

print("\nAfter Scaling")
print(scaled_df.min())
print(scaled_df.max())
```

```python
import pandas as pd

from sklearn.tree import DecisionTreeRegressor
from sklearn.feature_selection import RFE

df = pd.read_csv("/content/sample_data/zomato(1).csv")

X = df.drop("average_cost_for_two", axis=1)

y = df["average_cost_for_two"]

model = DecisionTreeRegressor(random_state=42)

rfe = RFE(model, n_features_to_select=3)

rfe.fit(X, y)

print("Selected Features")

print(X.columns[rfe.support_])

print("\nRanking")

for feature, rank in zip(X.columns, rfe.ranking_):
    print(feature, ":", rank)
```

```python
import pandas as pd

df = pd.read_csv("/content/sample_data/bookmyshow(1).csv")

Q1 = df["user_ratings"].quantile(0.25)

Q3 = df["user_ratings"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR

upper = Q3 + 1.5 * IQR

filtered = df[
    (df["user_ratings"] >= lower) &
    (df["user_ratings"] <= upper)
]

print("Original Rows:", len(df))

print("Rows After Removing Outliers:", len(filtered))

print("Dropped Rows:", len(df) - len(filtered))
```
