```python
import numpy as np
from sklearn.metrics import mean_squared_error

# Actual delivery times (minutes)
actual = np.array([25, 30, 28, 35, 40, 22, 27, 33, 38, 29])

# Predicted delivery times (minutes)
predicted = np.array([24, 31, 30, 34, 39, 23, 28, 32, 37, 30])

# Mean Squared Error
mse = mean_squared_error(actual, predicted)

# Root Mean Squared Error
rmse = np.sqrt(mse)

print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, r2_score

# Actual product prices
actual_price = np.array([500, 700, 900, 1200, 1500, 1800])

# Predicted prices
predicted_price = np.array([520, 680, 880, 1180, 1490, 1820])

# Function
def evaluate_model(actual, predicted):
    mae = mean_absolute_error(actual, predicted)
    r2 = r2_score(actual, predicted)

    print("Mean Absolute Error (MAE):", mae)
    print("R² Score:", r2)

# Call function
evaluate_model(actual_price, predicted_price)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Actual ratings (20 movies)
actual = np.array([
    4.5, 4.2, 3.8, 4.7, 3.5,
    4.0, 4.8, 3.9, 4.3, 4.6,
    3.7, 4.1, 4.9, 3.6, 4.4,
    4.0, 3.8, 4.5, 4.2, 4.7
])

# Predicted ratings
predicted = np.array([
    4.4, 4.1, 3.9, 4.6, 3.6,
    4.1, 4.7, 4.0, 4.2, 4.5,
    3.8, 4.0, 4.8, 3.7, 4.3,
    4.1, 3.9, 4.4, 4.3, 4.6
])

# Residuals
residuals = actual - predicted

# Residual Plot
plt.figure(figsize=(7,5))
plt.scatter(predicted, residuals, color="blue")
plt.axhline(y=0, color="red", linestyle="--")

plt.xlabel("Predicted Ratings")
plt.ylabel("Residuals")
plt.title("Residual Plot")

plt.grid(True)
plt.show()
```

```python
import numpy as np
from sklearn.metrics import (
    mean_squared_error,
    mean_absolute_error,
    r2_score
)

# Actual popularity
actual = np.array([90, 85, 70, 65, 80, 95, 60, 75])

# Predictions from Model 1
model1 = np.array([91, 84, 72, 64, 79, 94, 61, 76])

# Predictions from Model 2
model2 = np.array([95, 78, 65, 70, 82, 88, 55, 80])

def evaluate(actual, predicted):
    mse = mean_squared_error(actual, predicted)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(actual, predicted)
    r2 = r2_score(actual, predicted)

    return mse, rmse, mae, r2

m1 = evaluate(actual, model1)
m2 = evaluate(actual, model2)

print("Model 1")
print("MSE :", m1[0])
print("RMSE:", m1[1])
print("MAE :", m1[2])
print("R²  :", m1[3])

print("\nModel 2")
print("MSE :", m2[0])
print("RMSE:", m2[1])
print("MAE :", m2[2])
print("R²  :", m2[3])
```
