```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv("/content/sample_data/ipl_data.csv")

# Fill missing values (optional)
df["player_role"] = df["player_role"].fillna("Unknown")

# Create LabelEncoder object
encoder = LabelEncoder()

# Encode player_role
df["player_role_encoded"] = encoder.fit_transform(df["player_role"])

# Display DataFrame
print(df[["player_role", "player_role_encoded"]])

# Display mapping
print("\nRole Mapping:")

for role, value in zip(encoder.classes_, range(len(encoder.classes_))):
    print(role, "->", value)
```
